```r
# Generamos 100 calificaciones con una distribución normal
# media = 7.5, desviación estándar = 1.5
calificaciones <- rnorm(100, mean = 7.5, sd = 1.5)

# Redondeamos las calificaciones a enteros para simular calificaciones reales
calificaciones <- round(calificaciones)

# El resto del código se mantiene igual...
# (Calculamos medidas de tendencia central, dispersión, etc.)
# ...

# Calculamos las medidas de tendencia central: media, mediana y moda
media <- mean(calificaciones)
mediana <- median(calificaciones)

# Para la moda, necesitamos una función adicional:
getmode <- function(v) {
  uniqv <- unique(v)
  uniqv[which.max(tabulate(match(v, uniqv)))]
}
moda <- getmode(calificaciones)

# Calculamos las medidas de dispersión: rango, varianza y desviación estándar
rango <- max(calificaciones) - min(calificaciones)
varianza <- var(calificaciones)
desviacion_estandar <- sd(calificaciones)

# Imprimimos los resultados
cat("Medidas de tendencia central:\n")
cat("Media:", media, "\n")
cat("Mediana:", mediana, "\n")
cat("Moda:", moda, "\n")

cat("\nMedidas de dispersión:\n")
cat("Rango:", rango, "\n")
cat("Varianza:", varianza, "\n")
cat("Desviación estándar:", desviacion_estandar, "\n")

# Creamos el histograma
hist(calificaciones, main = "Histograma de Calificaciones", xlab = "Calificaciones", ylab = "Frecuencia", breaks = seq(min(calificaciones) - 0.5, max(calificaciones) + 0.5, by = 1), probability = TRUE) # Ajustamos breaks y probability para una mejor visualización

# Agregamos una curva de densidad para comparar con la distribución normal
curve(dnorm(x, mean = mean(calificaciones), sd = sd(calificaciones)), add = TRUE, col = "blue")
```
